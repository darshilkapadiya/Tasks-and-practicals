```python
class Calc:
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a - b
    def mul(self,a,b):
        return a*b 
    def div(self,a,b):
        return a/b
    def mod(self,a,b):
        return a%b

a = int(input("Enter Value of a"))
b = int(input("Enter Value of b"))

cal = Calc()

sel = input("Select one option 1. Add 2. Sub 3. Multiply 4. Divide 5. Modulo")
    
if sel == '1':
    print(f"Addition of {a} and {b}-->",cal.add(a,b))
elif sel == '2':
    print(f"Subtraction of {a} and {b}-->",cal.sub(a,b))
elif sel == '3':
    print(f"Multiplication of {a} and {b}-->",cal.mul(a,b))
elif sel == '4':
    print(f"Division of {a} and {b}-->",cal.div(a,b))
elif sel == '5':
    print(f"Modulo of {a} and {b}-->",cal.mod(a,b))
else:
    print("Invalid choice")
```
