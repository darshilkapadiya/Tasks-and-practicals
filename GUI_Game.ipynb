```python
from tkinter import *
import tkinter.messagebox as msg
```

```python
import random
```

```python
rps = Tk()
rps.title("Rock Paper Scissor Game")
rps.geometry("800x600")
rps.configure(bg="skyblue")
o = StringVar()
cm = StringVar()
point = StringVar()
points = StringVar()
re = StringVar()
count = 0
c = 0
def rockinp():
    global count
    global c
    o.set("Rock")
    cm.set(random.choice(['Rock','Paper','Scissor']))
    if cmp['text'] == "Scissor":
        re.set("User Wins and Computer Losses")
        count += 1
        point.set(count)
    elif cmp['text'] == "Rock":
        re.set("Draw Match")
    else:
        re.set("Computer Wins and user Losses")
        c += 1
        points.set(c)
    if count == 15:
        msg.showinfo("Success","You win")
    elif c == 15:
        msg.showinfo("Try Again","You lose")
    elif count == 15 and c == 15: 
        msg.showinfo("Retry","Draw Match")
    else:
        pass
def paperinp():
    global count
    global c
    o.set("Paper")
    cm.set(random.choice(['Rock','Paper','Scissor']))
    if cmp['text'] == "Rock":
        re.set("User Wins and computer Losses")
        count += 1
        point.set(count)
    elif cmp['text'] == "Paper":
        re.set("Draw Match")
    else:
        re.set("Computer Wins and user Losses")
        c += 1
        points.set(c)
    if count == 15:
        msg.showinfo("Success","You win")
    elif c == 15:
        msg.showinfo("Try Again","You lose")
    elif count == 15 and c == 15: 
        msg.showinfo("Retry","Draw Match")
    else:
        pass
    
def scinp():
    global count
    global c
    o.set("Scissor")
    cm.set(random.choice(['Rock','Paper','Scissor']))
    if cmp['text'] == "Paper":
        re.set("User Wins and computer Losses")
        count += 1
        point.set(count)
    elif cmp['text'] == "Scissor":
        re.set("Draw Match")
    else:
        re.set("Computer Wins and User Losses")
        c += 1
        points.set(c)
    if count == 15:
        msg.showinfo("Success","You win")
    elif c == 15:
        msg.showinfo("Try Again","You lose")
    elif count == 15 and c == 15: 
        msg.showinfo("Retry","Draw Match")
    else:
        pass
        
        
inp = Label(rps,text="Click any labeled button for input",bg="cyan",fg="darkblue",width=60,font=("Courier", 18))
#inp.config()
inp.place(x =0,y=0)

rock = Button(rps,text="The Rock",bg = "Blue",fg="white",width=20,command=rockinp)
rock.place(x = 50,y = 50)
paper = Button(rps,text="Paper",bg = "Blue",fg="white",width=20,command=paperinp)
paper.place(x = 300,y = 50)
scissor = Button(rps,text="Scissor",bg = "Blue",fg="white",width=20,command=scinp)
scissor.place(x = 600,y = 50)


outlab = Label(rps,text="User's Input",bg="white",fg="blue",font=("Courier", 10))
outlab.place(x = 50,y =170)

out = Button(rps,text="Output",bg = "Blue",fg="white",width=20,textvariable=o)
out.place(x = 50,y = 200)

cmplab = Label(rps,text="Computer's Input",bg="white",fg="blue",font=("Courier", 10))
cmplab.place(x = 600,y = 170)

cmp = Button(rps,text="Computer",bg = "Blue",fg="white",width=20,textvariable=cm)
cmp.place(x = 600,y = 200)

Result = Button(rps,text="Result",bg = "Cyan",fg="darkblue",width=50,textvariable=re)
Result.place(x = 200,y = 400)
 
userla = Label(rps,text="User points-->")
userla.place(x = 50,y = 400)

upoint = Button(rps,text="0",bg="Blue",fg="white",textvariable=point,width=8)
upoint.place(x = 150,y = 400)

cmpla = Label(rps,text="Machine points-->")
cmpla.place(x = 600,y = 400)

cpoint = Button(rps,text="0",bg="Blue",fg="white",textvariable=points,width=8)
cpoint.place(x = 700,y = 400)


rps.mainloop()
```
