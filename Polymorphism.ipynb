```python
class A:
  def display(self,a,b):
    self.a = a
    self.b = b
    print(f"Value of a is {self.a} and b is {self.b}")
  
  def display(self,a,b,c):
    self.a = a
    self.b = b
    self.c = c
    print(f"Value of a is {self.a} , b is {self.b} and c is {self.c}")
o = A()
o.display(11,12,30)
```

```python
class X:
  def display(self,a,b):
    self.a = a
    self.b = b
    print(f"Value of a is {self.a} and b is {self.b}")
class Y(X):
  def display(self,a,b):
    self.a = a
    self.b = b
    self.c = 33
    print(f"Value of a is {self.a} , b is {self.b} and c is {self.c}")
o = Y()
o.display(11,12)
```
