```python
def calc():   
    import tkinter

    win = tkinter.Tk()

    win.geometry("800x600")

    win.title("Calculater")

    lbl = tkinter.Label(win, text="Enter value in both boxes to perform any operation->")

    lbl.place(x=10,y=20)

    txt = tkinter.Entry(win)
    txt.place(x=300,y=20)

    txt1 = tkinter.Entry(win)
    txt1.place(x=400,y=20)

    def addi():
        a = txt.get()
        b = txt1.get()
        empty['text'] = f"{a} + {b}->",int(a)+int(b)
    def subt():
        a = txt.get()
        b = txt1.get()
        empty['text'] = f"{a} - {b}->",int(a)-int(b)
    def mult():    
        a = txt.get()
        b = txt1.get()
        empty['text'] = f"{a} * {b}->",int(a)*int(b)
    def divi():    
        a = txt.get()
        b = txt1.get()
        empty['text'] = f"{a} / {b}->",int(a)/int(b)
    def modu():    
        a = txt.get()
        b = txt1.get()
        empty['text'] = f"{a} % {b}->",int(a)%int(b)
    def event_add(e):
            addi()

    def event_sub(e):
            subt()
    def event_mul(e):
            mult()

    def event_div(e):
            divi()
    def event_mod(e):
            modu()  
    add = tkinter.Button(win,text="Addition",command = addi)
    add.place(x=300,y=40)

    sub = tkinter.Button(win, text="Subraction",command= subt)
    sub.place(x=(70+290),y=40)

    mul = tkinter.Button(win, text="Multiply",command=mult)
    mul.place(x=300,y=80)

    div = tkinter.Button(win, text="Division",command=divi)
    div.place(x=(70+290),y=80)

    mod = tkinter.Button(win, text="Modulo",command=modu)
    mod.place(x=(30+290),y=120)

    empty = tkinter.Label(win,text="Result")
    empty.place(x=320,y=180)
    win.bind("+",event_add)
    win.bind("-",event_sub)
    win.bind("*",event_mul)
    win.bind("/",event_div)
    win.bind("%",event_mod)
    

    win.mainloop()
```

```python
def incdec():    
    import tkinter

    win = tkinter.Tk()

    win.geometry("500x600")

    win.title("increment/decrement")

    def inc():
        zero['text']+= 1

    def dec():
        if zero['text'] == 0:
            zero['text'] = 0
        else:
            zero['text']-= 1

    def event_handle(e):
            inc()

    def event_handle1(e):
            dec()
    sub = tkinter.Button(win,text="-",command=dec)
    sub.place(x=220,y=200)

    zero = tkinter.Label(win,text=0)
    zero.place(x=250,y=200)

    add = tkinter.Button(win,text="+",command=inc)
    add.place(x=280,y=200)

    win.bind("+",event_handle)
    win.bind("-",event_handle1)
    win.mainloop()
```

```python
calc()
```
