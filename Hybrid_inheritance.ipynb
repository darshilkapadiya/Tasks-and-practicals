```python
# Simple example of hybrid inheritance no 1
class A:
  def method_a():
    print("Class A method called")
class B(A):
  def method_b():
    print("Class B method called")
class C(A):
  def method_c():
    print("Class c method called")
class D(A):
  def method_d():
    print("Class D method called")
class E(B,C,D):
  def method_e():
    print("Class e method called")
```

```python
o = E()
```

```python
spec(o)
```

```python
# Simple example of hybrid inheritane no 2
class X:
  def method_x():
    print("Class x method called")
class Y:
  def method_y():
    print("Class y method called")
class Z(X,Y):
  def method_z():
    print("Class Z method called")
class U(Z):
  def method_u():
    print("Class u method called")
class V(Z):
  def method_v():
    print("Class v method called")
```

```python
o1 = U()
o2 = V()
```

```python
spec(o1)
```

```python
spec(o2)
```

```python
def spec(dirname):
    l = list()
    l = dir(dirname)
    l1 = list()
    for i in l:
        if not i.startswith('__') or not i.endswith('__'):
            print(i)
```

```python
# Real life illustration 1 of hybrid inheritance 

class Dep():
  def method_dep(self):
    self.dep_name = 'Information technology' 
    print("Class Department method called")
    print(self.dep_name)
class Stu1(Dep):
  def method_stu1(self):
    self.name1 = 'abc'
    print("Class Stu1 method called")
    print("Name of student 1 and department-->",self.name1,self.dep_name)

class Stu2(Dep):
  def method_stu2(self):
    self.name2 = 'xyz'
    print("Class Stu2 method called")
    print("Name of student 2 and department-->",self.name2,self.dep_name)

class Stu3(Dep):
  def method_stu3(self):
    self.name3 = 'jamesbond'
    print("Class Stu3 method called")
    print("Name  of student 3 and department-->",self.name3,self.dep_name)

class Sem(Stu1,Stu2,Stu3):
  def method_sem(self):
    self.sem = '7th sem'
    Dep.method_dep(self)
    Stu1.method_stu1(self)
    Stu2.method_stu2(self)
    Stu3.method_stu3(self)
    print("Class Semester method called")
    print("Name of student 1 semester and department-->",self.name1,self.dep_name,self.sem)
    print("Name of student 2 semester and department-->",self.name2,self.dep_name,self.sem)
    print("Name of student 3 semester and department-->",self.name3,self.dep_name,self.sem)
```

```python
s = Sem()
```

```python
spec(s)
```

```python
s.method_sem()
```

```python
# Illustration no 2
class Dep:
    def method_dep(self):
      print("Class department method called")
class Proff1(Dep):
    def method_proff1(self):
      print("Class Proffesser1 method called ")
class Proff2(Dep):
    def method_proff2(self):
      print("Class Proffesser2 method called ")
class Proff3(Dep):
    def method_proff3(self):
      print("Class Proffesser3 method called ")
class Hod(Proff1,Proff2,Proff3):
    def method_hod(self):
      Dep.method_dep(self)
      Proff1.method_proff1(self)
      Proff2.method_proff2(self)
      Proff3.method_proff3(self)
      print("Class HOD method called")
```

```python
h = Hod()
```

```python
h.method_hod()
```

```python
# illustration no 3
class Grandparent:
  def method_g(self):
    print("Class grandparent method called")
class Father(Grandparent):
  def method_f(self):
    print("Class father method called")
class Mother(Grandparent):
  def method_m(self):
    print("Class mother method called")
class Child(Father,Mother):
  def method_c(self):
    Grandparent.method_g(self)
    Father.method_f(self)
    Mother.method_m(self)
    print("Class child method called")
```

```python
c = Child()
```

```python
c.method_c()
```
